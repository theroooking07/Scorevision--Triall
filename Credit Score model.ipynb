```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("credit_score_dataset_final.csv")
```

```python
data
```

```python
data
```

# Assigning a value to each feature

```python
income_mapping = {
    'Below ₹3,00,000': 300,
    '₹3,00,001 – ₹6,00,000': 400,
    '₹6,00,001 – ₹12,00,000': 500,
    '₹12,00,001 – ₹25,00,000': 600,
    'Above ₹25,00,000': 700
}
```

```python
data['Income_lvl']=data['Income'].map(income_mapping)
```

```python
data
```

```python
age_group_mapping = {
    '18–25': 300,
    '26–35': 400,
    '36–50': 500,
    '51–65': 400,
    'Above 65': 300
}
```

```python
data['Age_lvl']=data['Age Group'].map(age_group_mapping)
```

```python
data
```

```python
dependents_mapping = {
    0: 600,
    1: 550,
    2: 500,
    3: 450,
    4: 400,
    5: 350,
    6: 300
}
```

```python
data['Dep_lvl']=data['Number of Dependents'].map(dependents_mapping)
```

```python
data
```

```python
debts_mapping = {
    'Below ₹1,00,000': 700,
    '₹1,00,001 – ₹5,00,000': 600,
    '₹5,00,001 – ₹10,00,000': 500,
    '₹10,00,001 – ₹25,00,000': 400,
    'Above ₹25,00,000': 300
}
```

```python
data['Debts_lvl']=data['Debts'].map(debts_mapping)
```

```python
data
```

```python
education_mapping = {
    'High School': 300,
    'Diploma': 400,
    'Bachelors': 500,
    'Masters': 600,
    'PhD': 700
}
```

```python
data['Educ_lvl']=data['Education'].map(education_mapping)
```

```python
data
```

```python
assets_mapping = {
    'Below ₹2,00,000': 300,
    '₹2,00,000 – ₹10,00,000': 400,
    '₹10,00,001 – ₹50,00,000': 500,
    '₹50,00,001 – ₹1,00,00,000': 600,
    'Above ₹1,00,00,000': 700
}
```

```python
data['Assets_lvl']=data['Assets'].map(assets_mapping)
```

```python
data
```

```python
insurance_mapping = {
    'No missed payments': 600,
    '1–2 late payments': 400,
    'More than 2 missed payments': 300
}
```

```python
data['Insur_lvl']=data['Insurance Status'].map(insurance_mapping)
```

```python
data
```

# Splitting the data into 2 groups training and testing

```python
train_size=int(0.6*len(data))
```

```python
train_data=data[:train_size]
```

```python
test_data=data[train_size:]
```

```python
train_data
```

```python
test_data
```

# Scikit Learn Model Training

```python
credit_model=LinearRegression()
```

```python
credit_model.fit(train_data[['Income_lvl','Age_lvl','Dep_lvl','Debts_lvl','Educ_lvl','Assets_lvl','Insur_lvl']],train_data['Credit Score (Normalized)'])
```

# Predicted Credit Score for random value

```python
credit_model.predict([[400,500,350,500,300,400,300]])
```

# Predicted Credit Score for test data

```python
test_data['predicted_price']=credit_model.predict(test_data[['Income_lvl','Age_lvl','Dep_lvl','Debts_lvl','Educ_lvl','Assets_lvl','Insur_lvl']])
```

```python
test_data
```

```python
test_data['predicted_price']=test_data['predicted_price'].astype(int)
```

# Classification error Calculation

```python
from sklearn.metrics import classification_report
```

```python
r2 = r2_score(test_data['Credit Score (Normalized)'], test_data['predicted_price'])
```
